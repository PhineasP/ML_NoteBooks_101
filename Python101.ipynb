```python
def ConvertToM(x):
    megabyte = float(0.000976562)
    if x.find('M') != -1:
        return float(x.rstrip('M'))
    else:
        return float(x.rstrip('k'))*megabyte
```

```python
# play_store_df.loc[play_store_df['Size'] == 'Varies with device']


# Not useful
play_store_df = play_store_df.copy()[play_store_df['Size'] != '1,000+']
```

```python
play_store_df['Size'] = play_store_df['Size'].map(lambda x: 'Varies with device' if x == 'Varies with device' else ConvertToM(x) )
```

```python
hotencode_category = LabelEncoder()
play_store_df['category_encoded'] = hotencode_category.fit_transform(play_store_df['Category'])
```

```python
vwd_keys = play_store_df.loc[play_store_df['Size'] == 'Varies with device']['category_encoded'].unique()
for key in vwd_keys:
    temp_data = statistics.median(play_store_df[(play_store_df['category_encoded'] == key) & (play_store_df['Size'] != 'Varies with device')]['Size'])
    play_store_df.loc[((play_store_df['category_encoded'] == key) & (play_store_df['Size'] == 'Varies with device')), ['Size']] = temp_data
    
play_store_df['Size'] = play_store_df['Size'].map(lambda x: float(x))
```
